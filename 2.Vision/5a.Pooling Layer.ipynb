## Pooling Layer
- In general, pooling process is added to the pre-processing stage of image recognition to reduce the amount of inputs.
- Simply, pooling reduces the number of pixel by reducing its resolution.
    - It combine the area of pixel to single pixel. For this example, tf.nn.max_pool will be used which substitute the area to the single pixel with the maximum value.

### Example Dataset (MNIST)

```python
import keras
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
t = y_train
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

### Dataset Viasualize

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(11,4))

for i in range(30):
    subplot = fig.add_subplot(3, 10, i+1)
    subplot.set_xticks([])
    subplot.set_yticks([])
    subplot.set_title('%d' % t[i])
    subplot.imshow(x_train[i].reshape((28,28)), vmin=0, vmax=1,
                   cmap=plt.cm.gray_r, interpolation='nearest')
```

### Pooling with Keras

```python
import keras
from keras import Sequential
from keras.layers import MaxPooling2D
from keras import backend as K
```

```python
model = Sequential()
model.add(MaxPooling2D(pool_size=(3, 3),
                             input_shape=(28,28,1)
                             ))
```

```python
model.get_config()
```

```python
result_image = model.predict(x_train[:30].reshape(30,28,28,1))
```

### Visualize Result

```python
# Pool (3,3)
fig = plt.figure(figsize=(11,4))

for i in range(30):
    subplot = fig.add_subplot(3, 10, i+1)
    subplot.set_xticks([])
    subplot.set_yticks([])
    subplot.set_title('%d' % t[i])
    subplot.imshow(result_image[i,:,:,0], vmin=0, vmax=1,
                   cmap=plt.cm.gray_r, interpolation='nearest')
```
